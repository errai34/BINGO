```python
data = Table.read('/Users/ioanaciuca/Desktop/BINGO_P3/data/SD_pred_highquality.fits')
```

```python
mask = ((data['M_H_1'] > -1.0) & (data['R_2'] >=7.0) & (data['R_2'] <= 9.0) & (data['z_2c'] <= 2.0))
data_cut = data[mask]
```

```python
old_hal_mask = ((data_cut['ALPHA_M_1'] >= 0.12) & (data_cut['pred_logAge'] > 1.0))
old_hal = data_cut[old_hal_mask]
```

```python
young_hal_mask = ((data_cut['ALPHA_M_1'] >= 0.12) & (data_cut['pred_logAge'] < 0.8))
young_hal = data_cut[young_hal_mask]
```
