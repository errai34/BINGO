```python
import matplotlib
from astroML.crossmatch import crossmatch
from astroML.plotting.tools import draw_ellipse
from astroML.decorators import pickle_results
from astroML.stats import sigmaG

from astroML.plotting import setup_text_plots
setup_text_plots(fontsize = 14, usetex = False)
from astropy.table import Table
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

import sys
sys.path.append("/Users/ioanaciuca/Desktop/MasterFolder/Gaia_DR2/XD_tests/XDGMM/xdgmm/")
sys.path.append("/Users/ioanaciuca/Desktop/MasterFolder/Gaia_DR2/XD_tests/extreme-deconvolution/py/")
from xdgmm import XDGMM

import warnings
warnings.filterwarnings('ignore')
```

```python
plxlim = -1000.0
dataname = "./gdr2_files/gdr2_halo3.fits" 

data = Table.read(dataname, format = "fits") 
print "Length of raw data is:", len(data)

distance = float(raw_input("Insert distance: \n >>"))

print "distance chosen:", distance
print "Will now do the cuts"

plx = data["parallax"]
e_plx = data["parallax_error"]
gmag = data["phot_g_mean_mag"]

if distance >= 10:

	thr = 0.1
	k = 1
	data = data[(plx - k*e_plx <= thr) & (plx>plxlim)]

elif distance < 10:

	# data = data[(1.0/distance - data["parallax_error"] < data["parallax"]) & (data["parallax"] < 1.0/ distance + data["parallax_error"]) & (data["phot_g_mean_mag"] <= 17.0)]
        if distance < 2.0:
            print ' distance range between ',distance,' and ',distance+distance,' kpc and G<17 mag'
            data = data[(plx<1.0/distance) & (plx>1.0/(distance+distance)) & (gmag<=17.0)]
        elif distance < 3.0:
            print ' distance range between ',distance,' and ',distance+distance,' kpc and G<18 mag'
            data = data[(plx<1.0/distance) & (plx>1.0/(distance+distance)) & (gmag<=18.0)]
        elif distance < 4.0:
            # for 3- 6
            print ' distance range between ',distance,' and ',distance+distance,' kpc and G<19 mag'
            data = data[(plx<1.0/distance) & (plx>1.0/(distance+distance)) & (gmag<=19.0)]
        else:
            print ' distance range between ',distance,' and ',distance+distance,' kpc'
            # for 4-8, 5-10, 7-14, 9-18
            data = data[(plx<1.0/distance) & (plx>1.0/(distance+distance))]

print "Length of data after cuts: ", len(data)

par = np.array(data["parallax"]) # Angle[mas]
epar = np.array(data["parallax_error"])

pmra = np.array(data["pmra"]) # proper motion in mas/year
epmra = np.array(data["pmra_error"])

pmdec = np.array(data["pmdec"])
epmdec = np.array(data["pmdec_error"])

pmrapmdecor = np.array(data["pmra_pmdec_corr"])

glon = np.array(data["l"])
glat = np.array(data["b"])

eglon = 0.01 * np.ones_like(glon)
eglat = 0.01 * np.ones_like(glat)

X = np.vstack([pmra, pmdec, glon, glat]).T
Xerr = np.zeros(X.shape + X.shape[-1:])
diag = np.arange(X.shape[-1])
Xerr[:, diag, diag] = np.vstack([epmra ** 2, epmdec ** 2, eglon ** 2, eglat ** 2]).T
Xerr[:, 0, 1] =  pmrapmdecor * epmra * epmdec 
Xerr[:, 1, 0] =  pmrapmdecor * epmra * epmdec 
```

```python
xdgmm = XDGMM(method = 'Bovy')

n_components = int(raw_input("Insert number of components: \n >>"))

pickle_name = "./new_pkl_files/source_id3_dist%s_ncomp%s.pkl" % (int(distance), int(n_components))

@pickle_results(pickle_name)

def do_xdgmm(clf = xdgmm, X = X, Xerr = Xerr):
    
    clf.n_components = n_components
    clf.fit(X, Xerr)

    return clf

clf = do_xdgmm(clf = xdgmm, X = X, Xerr = Xerr)
```

```python
X_sample = clf.sample(X.shape[0])

import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(2*6.5, 6.5))

ax1.scatter(X[::3, 0], X[::3, 1], s=5, lw=0, c='k')

   
#ax2.scatter(X_sample[::5, 0], X_sample[::5, 1], s=9, lw=0, c='k')

# print "clf", clf.mu[0, 0:2]
# print "V", clf.V[0, 0:2, 0:2]

# sourt with length of principle component
sortindx = np.argsort(clf.V[:,0,0])
sortindx = sortindx[::-1]
for i in range(clf.n_components):
    ic = sortindx[i]
    print i,' component mu, C =',clf.mu[ic,0:2], clf.V[ic, 0:2, 0:2], \
      ' l,b mu, C=',clf.mu[ic, 2:4], clf.V[ic, 2:4, 2:4]
    draw_ellipse(clf.mu[ic, 0:2], clf.V[ic, 0:2, 0:2], scales=[2],
                 ec='k', fc='gray', alpha=0.2, ax=ax2)
        
    #ax3.scatter(pmra_dwarf, pmdec_dwarf, color = 'r', s = 1)

ax2.set_xlim(-10, 10)
ax2.set_ylim(-10, 10)

ax1.set_xlim(-10, 10)
ax1.set_ylim(-10, 10)

ax1.set_xlabel(r"$\mu_{\alpha^*} \rm (mas/yr)$", fontsize=18)

ax2.set_xlabel(r"$\mu_{\alpha^*}  \rm (mas/yr)$", fontsize=18)
ax1.set_ylabel(r"$\mu_{\delta} \ \rm (mas/yr)$", fontsize=18)
ax1.set_title("Data", fontsize=18)

ax2.set_title("XD cluster location", fontsize=18)

plt.savefig('/Users/ioanaciuca/Desktop/sagcandidate.pdf', format = 'pdf', dpi = 1000)
```
